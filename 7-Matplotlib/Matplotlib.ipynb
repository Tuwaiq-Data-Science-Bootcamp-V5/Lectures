# Data Visualization

How to represents the cleaned data in charts that depict patterns, trends, and correlations. So, it is a method for analyzing the data.

Many tools available for data visualization :-
1. By using Programming languages like Python
    - Matplotlib
    - Seaborn
    - Plotly
    - ggplot2
2. By using ready to use tools like :-
    - Excel

# Types of Data

1. Data Types
    - Integers
    - Floats
    - Strings
    - Booleans
    - Datetimes

2. Feature Type
    - Quantitive
        Magnitude, Price, etc.
    - Qualitative
        Color, Gender, etc.
    - Ordinal
        Grade, Age, etc.

3. Continuous and Discrete Variables
    - Continuous :
        Uncountable data, Time takes to run a marathon.

    - Discrete :
        Countable data, Time in minutes.

3. Who is the end user and what do they need?

# 1.1 Bar Chart :
    - Also known as Bar plots or Column Plots/Charts.
    - It is a way to represents categorical data in rectangular bars.
    - The bars can be plotted in horizontal way or vertical way.
    - It is used to compare the values of different groups of variables.
    - The height or length of the bar corresponds to the frequency of the variable or one numerical variable.

#### Examples :-
1. How many people use labtops, phones, or desktops in the company?
2. What is the mean price of cars for each car category?

### Question 1

How many people use labtops, phones, or desktops in the company?

````python
df = pd.DataFrame({
    'Device' : [
        'Laptop' if i in range(1, 45+1) else 'Phone' if i in range(45+1, 45+70+1) else 'Desktop' for i in range(1, 145+1)
    ]
})

df
````

````python
df['Device'].value_counts()
````

````python
categories = list(df['Device'].value_counts().index)
counts     = list(df['Device'].value_counts().values)

plt.bar(
    x = categories,            # The x axes
    height = counts,           # The y axes
)

plt.show();
````

#### Other Parameters

````python
plt.bar(
    x = categories,
    height = counts,
    width = 0.75,              # The width of the bin
    color = ['red', 'blue'],   # The colors of the bins
    edgecolor = 'white',       # The edge color of the bins
)

# Set the title of the chart
plt.title('Number of People Using Laptops, Phones, and Desktops in the Company')
plt.xlabel('Device Categories') # Set the x axis title
plt.ylabel('Number of People')  # Set the y axis title

plt.grid() # Showing the grid of the plot

plt.show(); # Displays the chart
````

# 1.2 Clustered Column/Bar :

    - Same as Bar char, but with two or more y axes variables

````python
df = pd.DataFrame(
    {
        'Products' : ['Prod 1', 'Prod 2', 'Prod 3', 'Prod 4', 'Prod 5'],
        'y1'       : [34, 56, 12, 89, 67],
        'y2'       : [12, 56, 78, 45, 90],
        'y3'       : [14, 23, 45, 25, 89]
    }
)

width = 0.2

plt.bar(np.arange(len(df))-width, df['y1'], width, color='cyan')
plt.bar(np.arange(len(df)),       df['y2'], width, color='orange')
plt.bar(np.arange(len(df))+width, df['y3'], width, color='green')

plt.xticks(np.arange(len(df)), list(df['Products'].value_counts().index))

plt.xlabel("Products")
plt.ylabel("Benifits")
plt.legend(["y1", "y2", "y3"])
plt.show();
````

# 1.3 Stacked Bar/Column Chart :

    - Same as clustered bar, but the series data are stacked on top of one another.

````python
plt.bar(np.arange(len(df)), df['y1'], color='r')
plt.bar(np.arange(len(df)), df['y2'], color='y')
plt.bar(np.arange(len(df)), df['y3'], color='g')

plt.xlabel("Products")
plt.ylabel("Benifits")
plt.legend(["y1", "y2", "y3"])
plt.show();
````

# 2.1 Line Chart :

    - Also known as line plot.
    - Represents data as a series of data points called markers.
    - Each marker connected by straight line segments.
    - Used for showing trends and patterns over time.

#### Example :-
- Show the relationship between the cost and benifits over months.

````python
benefits = [random.randint(10000, 50000) for i in range(12)]
months   = [datetime(2022, i, 1)         for i in range(1, 12+1)]

plt.plot(
    months, benefits,
    marker='o', linestyle='--',
    color='b', label='Benefits'
)

plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%b %Y'))
plt.gca().xaxis.set_major_locator(plt.matplotlib.dates.MonthLocator())
plt.gcf().autofmt_xdate()

plt.title('Monthly Benefits in 2022', fontsize=15)
plt.xlabel('Months', fontsize=12)
plt.ylabel('Benefits ($)', fontsize=12)
plt.legend() # Set the legend of the chart

plt.show();
````

````python
costs    = [random.randint(5000, 30000)  for i in range(12)]
benefits = [random.randint(10000, 50000) for i in range(12)]

months = [datetime(2022, i, 1) for i in range(1, 13)]

plt.plot(months, costs, marker='o', linestyle='--', color='r', label='Costs')
plt.plot(months, benefits, marker='o', linestyle='--', color='b', label='Benefits')

plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%b %Y'))
plt.gca().xaxis.set_major_locator(plt.matplotlib.dates.MonthLocator())
plt.gcf().autofmt_xdate()

plt.title('Monthly Costs and Benefits in 2022', fontsize=15)
plt.xlabel('Months', fontsize=12)
plt.ylabel('Amount ($)', fontsize=12)

plt.legend()

plt.show();
````

# 2.2 Area Chart : 

    - Also known as stacked area plot.
    - It is a chart that visualize quantitive data over time.
    - Same as line plot, But with area between the line and axis.
    - Use area plot when multiple aspects of data need to be shown.

````python
years = [2010, 2011, 2012, 2013, 2014, 2015, 2016]
categories = ['A', 'B', 'C', 'D', 'E']

df = pd.DataFrame({
    'date' : [2017, 2018, 2019, 2020, 2021, 2022, 2023],
    'a'    : [5, 4, 6, 4, 5, 6, 7],
    'b'    : [3, 2, 3, 4, 3, 4, 5],
    'c'    : [1, 2, 2, 3, 3, 4, 4],
    'd'    : [4, 3, 4, 3, 2, 3, 4],
    'e'    : [2, 1, 2, 2, 1, 2, 3]
})

plt.stackplot(
    df['date'],
    [df['a'], df['b'], df['c'], df['d'], df['e']],
    labels=['a', 'b', 'c', 'd', 'e'], alpha=0.8
)

plt.title('Area Plot', fontsize=15)
plt.xlabel('Years', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.legend(loc='upper left')

plt.show();
````

# 3.1 Scatter Plot :

    - It is a chart that represents the data as points.
    - Each point has a value from two variables, one from the x axes and the other one from y axes.
    - It is used to understand the relationship or correlation between two continuous variables.
    - It helps to identify clusters, groups, outliers, and other useful patterns.
    - Linear Regression algorithm it depends on scatter plot.
    - In most cases, the y axis as a "dependent" variable, and the x axis as an "independent" variable.

````python
data = pd.read_csv('Housing.csv')

data
````

````python
data['price'] = data['price'] / 1000
````

````python
plt.scatter(
    data['area'], data['price'],
    color='b', alpha=0.5, label='Houses'
)

plt.title('Relationship Between House Area and Price', fontsize=15)
plt.xlabel('House Area (in sq. ft)', fontsize=12)
plt.ylabel('House Price (in $1000)', fontsize=12)
plt.legend()

plt.show();
````

# 3.2 Bubble Chart :

    - Same as scatter plot, but in third dimension.
    - The points represented as bubbles, each of them has a value from three variables.
    - The first two values are from the axeses (x axes and y axes).
    - The third variable is the size of the bubble.
    - Use scatter function same as before, but you need to use s parameter.

````python
x = np.random.rand(30)
y = np.random.rand(30)
sizes = np.random.rand(30) * 800
colors = np.random.rand(30)

plt.scatter(
    x, y, s=sizes,
    c=colors, alpha=0.7, cmap='viridis'
)

plt.title('Bubble Plot Example', fontsize=15)
plt.xlabel('X-axis', fontsize=12)
plt.ylabel('Y-axis', fontsize=12)
plt.colorbar(label='Color Intensity')

plt.show();
````

# 4 Pie Chart :

    - Circular graph divides data into slices to visualize numerical proportions.
    - Each slice in the pie chart represents a category.
    - Consider using pie chart to show the percentage of different categories that make up a whole.

````python
data = pd.read_csv('final_cars_datasets.csv')

mark = data['mark'].value_counts().sort_values(ascending=False)[:7]

plt.pie(
    mark, labels=mark.index,
    autopct='%1.1f%%', startangle=90
)

plt.title('Distribution of Used Cars by Brand', fontsize=15)

plt.show();
````

````python
data = pd.read_csv('final_cars_datasets.csv')

mark = data['mark'].value_counts().sort_values(ascending=False)[:7]

plt.pie(
    mark, labels=mark.index,
    autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors,
    shadow=True, radius=1, explode = [0.3,0.2,0,0,0,0,0]
)

plt.title('Distribution of Used Cars by Brand', fontsize=15)

plt.show();
````

# 5 Histogram :

    - Represents the distribution of a numerical variable.
    - It consists of series of bins that placed on the x axis.
    - The width of each bin represnts the range of the values.
    - The height of each bin represnts the frequency of the data points.
    - Histograms helps you to find a lot of information.
    - Such as : Patterns, variations, trends, central tendency, dispersion, skewness, etc.
    - Use the height of each bin to identify outliers in the data.

````python
np.random.seed(0)
ages = np.random.normal(40, 10, 1000)

plt.hist(ages, bins=30, color='skyblue', edgecolor='white')

plt.title('Histogram of Age', fontsize=15)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

plt.show();
````

````python
np.random.seed(0)
sales = np.random.normal(500, 100, 500)

plt.hist(sales, bins=20, color='lightgreen', edgecolor='black')

plt.title('Histogram of Sales', fontsize=15)
plt.xlabel('Sales', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

plt.show();
````

````python
np.random.seed(0)
aluminum_skewed = np.random.gamma(4, 0.1, 500) + 0.2

plt.hist(aluminum_skewed, bins=15, color='lightcoral', edgecolor='white')

plt.title('Skewed Histogram of Aluminum in Water', fontsize=15)
plt.xlabel('Aluminum Concentration', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

plt.show();
````

# 6 Box and Whisker :

    - Also known as Box Plot.
    - It is used to visualize the distribution of the data with respect to the summary statistical measures (minimum, first quartile (Q1), second quartile (Q2) or median, third quartile (Q3), and maximum).
    - Used to understand the state of the data.
    - Is it contains outliers? what is the central tendency and variability?
    - The "whiskers," or lines extending from the box, show the variability outside the upper and lower quartiles, typically covering 1.5 times the IQR. Data points beyond the whiskers are often considered outliers and are plotted individually.

````python
np.random.seed(0)
data = np.random.normal(0, 1, 100)

plt.boxplot(data, vert=False, patch_artist=True)
plt.yticks([])

plt.show();
````

````python
np.random.seed(0)
data = np.concatenate(
    [
        np.random.normal(0, 1, 100),
        np.array([5, 7, -3, -5])
    ]
)

plt.boxplot(data, vert=False, patch_artist=True)
plt.yticks([])

plt.show();
````

````python
np.random.seed(0)
skewed_data = np.random.exponential(scale=2, size=1000) + 5

plt.boxplot(skewed_data, vert=False, patch_artist=True)
plt.yticks([])

plt.show();
````

# 8 Figure and Axes :

    - Figure : The window that you plot on, it has a size, so everything ploted is inside that figure.
    - Axes : You can say it is the sub windows for each plot.

````python
# Create a figure
fig = plt.figure(figsize=(16, 8))

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label='sin(x)', color='b', linestyle='-', linewidth=2)

plt.title('Sine Wave Plot', fontsize=16)
plt.xlabel('X-axis', fontsize=14)
plt.ylabel('Y-axis', fontsize=14)

plt.legend()

plt.show();
````

````python
fig, ax = plt.subplots(figsize=(8, 8))

x = np.linspace(0, 10, 100)
y = np.sin(x)

ax.plot(x, y, label='sin(x)', color='b', linestyle='-', linewidth=2)

ax.set_title('Sine Wave Plot', fontsize=16)
ax.set_xlabel('X-axis', fontsize=14)
ax.set_ylabel('Y-axis', fontsize=14)

ax.legend()

plt.show();
````

# 9 Subplots :

````python
fig, axs = plt.subplots(nrows=3, ncols=2, figsize=(20, 20))

x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

axs[0, 0].plot(x, y1, color='r')
axs[0, 0].set_title('Sin Plot')

axs[0, 1].plot(x, y2, color='b')
axs[0, 1].set_title('Cos Plot')

axs[1, 0].plot(x, y1, color='g')
axs[1, 0].set_title('Sin Plot')

axs[1, 1].plot(x, y2, color='m')
axs[1, 1].set_title('Cos Plot')

axs[2, 0].plot(x, y1, color='c')
axs[2, 0].set_title('Sin Plot')

axs[2, 1].plot(x, y2, color='y')
axs[2, 1].set_title('Cos Plot')

plt.show();
````
