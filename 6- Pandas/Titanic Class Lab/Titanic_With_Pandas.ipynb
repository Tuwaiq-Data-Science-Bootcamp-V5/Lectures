```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("./titanic/train.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
```

```python
df["Age"] = df["Age"].astype(int)
```

```python
df.isnull().sum()
```

```python
df['Embarked'].mode()
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
df.drop("Cabin", axis= 1, inplace = True)
```

```python
df.head()
```

```python
df["Name"] = df["Name"].str.replace("(\w+(?=(\s|\.|$)))|,|\.|\s", "", regex= True)
```

```python
df.head()
```

```python
df.columns = ["PassengerId", "Survived", "Pclass", "Last_Name", "Sex", "Age", "SibSp", "Parch", "Ticket", "Fare", "Embarked"]
```

```python
df.head()
```

```python
df["PassengerId"] = df["PassengerId"].apply(lambda x: round(x/10))
```
