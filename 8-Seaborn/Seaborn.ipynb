# Seaborn

Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

# Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns

sns.set() # to set the style
```

# Read Database using SQLite Library

```python
import sqlite3

connection = sqlite3.connect('mpg.db') # miles per gallons
```

```python
query = 'SELECT * FROM mpg'
df = pd.read_sql(query, connection)
df.head()
```

```python
query2 = 'SELECT * FROM mpg where mpg = 15.0'
df2 = pd.read_sql(query2, connection)
df2.head()
```

```python
df.info() # horsepower 392 has a null
```

```python
df.shape #(rows, columns)
```

```python
# describe numerical columns
df.describe()
```

```python
# describe categorical columns
df.describe(exclude="number")
```

```python
# How many missing values
df.isnull().sum()
```

# Histogram

```python
# Histogram 
plt.figure(figsize=(15,7))

sns.histplot(df["mpg"])

plt.title("Example Histogram");
```

```python
# Histogram
plt.figure(figsize=(15,10))
sns.histplot(df["mpg"], bins=20) 
plt.title("Example Histogram");
```

```python
# Stacked Histogram separated by origin
plt.figure(figsize=(15,8))

sns.histplot(
    df,
    x="mpg",
    hue="origin", # color of plot elements
    multiple="stack", # also can be layer or dodge or fill
    alpha=0.8
)

plt.title("Example Stacked Histogram");
```

```python
sns.jointplot(x='mpg', y='weight',data=df, kind='scatter'); # kind can be'hex', 'kde' or 'reg'
```

# Pairplots

```python
# Remove unwanted columns
df_pair = df.drop(["index", "model_year"], axis=1)
sns.pairplot(df_pair, hue="origin", palette='coolwarm'); #for numerical values, we use hue for catagorical
```

# KDE Plots - kernel density estimate (KDE)

```python
sns.kdeplot(
    data=df, 
    x="mpg",
    hue="origin"
)

plt.title("Example KDE Plot");
```

# Barplot

```python
sns.barplot(x=df['origin'], y=df['horsepower'], estimator=np.mean); # it shows the average
```

```python
import numpy as np

sns.barplot(x='origin', y='horsepower',data=df, estimator=np.std);
```

# Countplot

```python
sns.countplot(x='origin', data=df); # it shows the count 
```

```python
categories_df = df[['origin', 'name', 'cylinders','model_year']]
categories_df.head()
```

```python
categories = list(df.select_dtypes(include=['object']).columns)
print('Categorical variables: ', categories)
numbers = list(df.select_dtypes(exclude=['object']).columns)
print('Numerical variables: ', numbers)
```

```python
categories_df['mpg_level'] = df['mpg'].apply(lambda x: 'Low' if x < 15 else 'High' if x > 30 else 'Medium')
categories.append('mpg_level')
print('Categorical variables: ', categories)
```

```python
plt.figure(figsize=(15,8))
sns.countplot(x = 'model_year', hue = 'mpg_level', data = categories_df);

# manufacturing of low mpg car decreases and eventually stopped after 79.
# manufacturing of high mpg car increases and infact after 79, 
# their manufacturing was 1.5-2 times as compared to vehicles with medium mpg.
# Throughout the years there is no significant change in manufacturing of vehicles with medium mpg and 
# infact from 74-79 their manufacturing was more than the sum of remaining two.
```

```python
sns.relplot(x = 'model_year', y = 'mpg', data = df)
plt.show();
# From the scatter plot we can clearly see an upward linear trend i.e., as year progress mpg increases.
```

# Grouped Bar Chart

```python
df['cylinders'] = df['cylinders'].astype(str)

g = sns.catplot(
    data=df, # Data
    kind="bar", # Chart to use
    x="origin", # X axis
    y="horsepower", # y axis
    hue="cylinders", # create bar for each hue
    alpha=.8, # adjust transparancy
    height=5
)

plt.title("Example Grouped Bar Chart");
```

# Boxplot

```python
sns.boxplot(df["mpg"]);
```

```python
df["mpg"].describe()
```

```python
sns.boxplot(x="origin", y="mpg", data=df, palette='rainbow');
```

```python
plt.figure(figsize=(15,8))

sns.boxplot(
    x="origin", 
    y="mpg",
    hue="cylinders",
    data=df)

plt.title("Example Grouped Boxplots");
```

```python
plt.figure(figsize=(15,8))
df['modren_models'] = df['model_year'] > 76
sns.boxplot(x="origin", y="mpg", hue='modren_models', data=df);
```

# Violin Plot

```python
# Draw a nested violinplot and split the violins for easier comparison
plt.figure(figsize=(10,8))

sns.violinplot(
    data=df, 
    x="origin", 
    y="mpg", 
    hue="cylinders",
    linewidth=1
)

plt.title("Sample Violin Plot");
```

# Swarm Plots

```python
plt.figure(figsize=(10,8)) # it does not scale well to large numbers of observations.

sns.swarmplot(
    x="origin",
    y="mpg", 
    hue="cylinders", 
    data=df
)

plt.title("Example Swarm Plot");
```

# Relational Scatter Plot

```python
# Plot miles per gallon against horsepower with other semantics

sns.relplot(
    x="horsepower", 
    y="mpg", 
    hue="origin",
    size="weight",
    alpha=.8, 
    palette="muted",
    height=6, 
    data=df
)

plt.title("Example Relational Scatter Plot");
```
